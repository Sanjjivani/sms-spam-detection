# 3.Data preprocessing 
1]lowercase 
2] Tokenization
3]removing special characters 
4}Removing stop word and puncation
5]Stemming                    

```python
import nltk
from nltk.corpus import stopwords
import string
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()

def transform_text(text):
    #lowercase
    text = text.lower()
    
    #tokeniziation
    text= nltk.word_tokenize(text)

    #removing special characters 
    y=[]
    for i in text: 
        if i.isalnum():
            y.append(i)    

    ##4}Removing stop word and puncation
    text = y[:]
    y.clear()

    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)

    text=y[:]
    y.clear()
    
 ##5]Stemming 
    for i in text:
        y.append(ps.stem(i))
            
    return " ".join(y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB  
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier  
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
# Support Vector Classifier with sigmoid kernel and gamma parameter
svc = SVC(kernel='sigmoid', gamma=1.0)  # Corrected 'kernal' to 'kernel' and 'sigmold' to 'sigmoid'

# K-Nearest Neighbors Classifier
knc = KNeighborsClassifier()

# Naive Bayes
mnb = MultinomialNB()

# Decision Tree Classifier with max depth
dtc = DecisionTreeClassifier(max_depth=5)

# Logistic Regression with liblinear solver and l2 penalty (corrected '11' to 'l2' for the penalty)
lrc = LogisticRegression(solver='liblinear', penalty='l2')

# Random Forest Classifier with 50 estimators and random state
rfc = RandomForestClassifier(n_estimators=50, random_state=2)

# AdaBoost Classifier with 50 estimators and random state
abc = AdaBoostClassifier(n_estimators=50, random_state=2)  # Moved 'random_state' inside

# Bagging Classifier with 50 estimators and random state
bc = BaggingClassifier(n_estimators=50, random_state=2)

# Extra Trees Classifier with 50 estimators and random state
etc = ExtraTreesClassifier(n_estimators=50, random_state=2)

# Gradient Boosting Classifier with 50 estimators and random state
gbdt = GradientBoostingClassifier(n_estimators=50, random_state=2)

# XGBoost Classifier with 50 estimators and random state
xgb = XGBClassifier(n_estimators=50, random_state=2)
```

```python
clfs = {
        'SVC' : svc,
        'KN' :knc,
        'NB' : mnb,
        'DT' : dtc,
        'LR' : lrc,
        'RF' : rfc,
        'AdaBoost': abc,
        'Bgc' : bc,
        'ETC' : etc,
        'GBDT' : gbdt,
        'xgb' : xgb

    
}
```

```python


def train_classifier(clf, X_train, y_train, X_test, y_test):
    # Train the classifier
    clf.fit(X_train, y_train)
    
    # Make predictions on the test data
    y_pred = clf.predict(X_test)
    
    # Calculate accuracy and precision
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    
    # Return accuracy and precision
    return accuracy, precision
```

```python
train_classifier(svc,X_train , y_train ,X_test , y_test)
```

```python
accuracy_scores = []
precision_scores = []

for name, clf in clfs.items():
    # Train the classifier and get accuracy and precision
    current_accuracy, current_precision = train_classifier(clf, X_train, y_train, X_test, y_test)
    
    # Print the results for the current classifier
    print("For ",name)
    print("Accuracy -",current_accuracy)
    print("Precision -",current_precision)
    
    # Append the results to the respective lists
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
```

```python
performance_df = pd.DataFrame({
    'Algorithm': clfs.keys(),
    'Accuracy': accuracy_scores,
    'Precision': precision_scores
})
```
